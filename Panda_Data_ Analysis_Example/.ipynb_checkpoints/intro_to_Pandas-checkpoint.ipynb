## Manipulating Data

```python
car_sales["Lower"] = car_sales["Make"].str.lower()
car_sales
```

```python
#En la vida real es normal no tener dataframes completos, tal:
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
car_sales_missing
```

```python
# fillna llena los NaN en el dataframe

#Mediante la sobreescritura:
# car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(0)

#Usando inplace = True se sustituye sobre el actual.
car_sales_missing["Odometer"].fillna(0, inplace=True)
car_sales_missing

#De esta forma se actualiza el dataframe con el cambio recien
```

```python
# dropna() elimina todo lo que posea NaN

car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
#Columns from series
seats_column = pd.Series([5, 5])
car_sales["Seats"] = seats_column

#Quitaré la columna Lower que había añadido antes
#car_sales.drop(columns = "Lower", inplace=True)
car_sales["Seats"].fillna(5, inplace=True)
car_sales
```

```python
#Columns from Python List

#Para hacer esto, la lista debe tener el mismo length que el dataframe
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel/100KM"] = fuel_economy
car_sales
```

```python
#Creando una nueva columna a partir de otra:
car_sales["Total Fuel Used"] = car_sales["Odometer (KM)"]/100 *car_sales["Fuel/100KM"]
car_sales
```

```python
#Crear una columna a partir de un solo valor:
car_sales["Wheels"] = 4
car_sales
```

```python
#Algo importante para el entrenamiento de modelos en ML, es
#que la data se debe randomizar para un mejor crecimiento.

#Podemos tomar 'samples' random de la data en proporción 0 a 1:
#  car_sales.sample(frac=0.4)
#Si usamos '1', se toma toda la data de manera aleatoria.
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

```python
#Se puede resetear los index para el orden actual:
car_sales_shuffled.reset_index(drop=True, inplace=True)
car_sales_shuffled

#Pero nótese que se añade una nueva columna 'index' con los 
#index anteriores. Evitamos esto con drop=True
```

```python
#Aplicar una función a una columna:

#car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x/2)
car_odometer = car_sales["Odometer (KM)"].apply(lambda x: x/2)
car_odometer
```
